```python
int_ind_con = only_international_conference[only_international_conference['Location'].str.contains('India')]
```

```python
int_ind_con.dropna(subset=['External URL'])
```

```python
int_ind_con = int_ind_con.dropna(subset=['External URL'])
```

```python
start_date = '2025-05-24'
end_date = '2025-07-24'
```

```python
int_ind_con[(int_ind_con['Start Date'] >= start_date) & (int_ind_con['Start Date'] <= end_date)]
```

```python
int_ind_con_before_may_25 = int_ind_con[(int_ind_con['Start Date'] >= start_date) & (int_ind_con['Start Date'] <= end_date)]
```

```python
int_ind_con_before_may_25_reset = int_ind_con_before_may_25.reset_index(drop=True)
```

```python
int_ind_con_before_may_25_reset
```

```python
int_ind_con_before_may_25_reset.sort_values(by='Start Date')
```

```python
final_df = int_ind_con_before_may_25_reset.sort_values(by='Start Date')
```

```python
final_df = final_df.reset_index(drop=True)
```
